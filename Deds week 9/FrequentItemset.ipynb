```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
import sqlite3
# Connect to the SQLite database
goStaff = sqlite3.connect('../go_sales.sqlite')

# Read product table into a DataFrame
product_df = pd.read_sql_query("SELECT * FROM product", goStaff)
orders_df = pd.read_sql_query("SELECT * FROM order_details", goStaff)

merged_df = pd.merge(orders_df, product_df, on='PRODUCT_NUMBER', how='left')

merged_df
```

```python
merged_df = merged_df.loc[:,["ORDER_NUMBER","PRODUCT_NAME"]]
merged_df
```

```python
transactions = merged_df.groupby('ORDER_NUMBER')['PRODUCT_NAME'].apply(list).tolist()
te = TransactionEncoder()
te_ary = te.fit_transform(transactions)
encoded_df = pd.DataFrame(te_ary, columns=te.columns_)
encoded_df
```

```python
apriori(encoded_df, min_support=0.01)
```

```python
frequent_itemsets = apriori(encoded_df, min_support=0.004, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] >= 4) &
                   (frequent_itemsets['support'] >= 0.004) ]
```

```python
import requests

# Fetch data from the API
response1 = requests.get('https://fakestoreapi.com/products')
response2 = requests.get('https://fakestoreapi.com/products/categories')

if response1.status_code == 200:
    products_data = response1.json()
    productsapi_df = pd.DataFrame(products_data)
    
if response2.status_code == 200:
    categories_data = response2.json()
    categoriesapi_df = pd.DataFrame(categories_data)
    categoriesapi_df = categoriesapi_df.rename(columns={0: 'category'})
    
df = pd.merge(categoriesapi_df, productsapi_df, on='category', how='left')
df
```

```python
df = df.loc[:,["category","title"]]
df
```

```python
transactions = df.groupby('category')['title'].apply(list).tolist()
te = TransactionEncoder()
te_ary = te.fit_transform(transactions)
encoded_df = pd.DataFrame(te_ary, columns=te.columns_)

encoded_df
```

```python
apriori(encoded_df, min_support=0.2)
```

```python
frequent_itemsets = apriori(encoded_df, min_support=0.004, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
frequent_itemsets[ (frequent_itemsets['length'] >= 4) &
                   (frequent_itemsets['support'] >= 0.2) ]
```
